```python
from transformers import pipeline

models = {
    "bert": "bert-base-uncased",
    "roberta": "roberta-base",
    "bart": "facebook/bart-base"
}
```

```python
prompt = "The future of Artificial Intelligence is"

for name, model_id in models.items():
    print(f"\n=== {name.upper()} ===")
    try:
        generator = pipeline("text-generation", model=model_id)
        output = generator(prompt, max_length=40, num_return_sequences=1)
        print(output[0]["generated_text"])
    except Exception as e:
        print("ERROR:", e)
```

```python
text = "The goal of Generative AI is to [MASK] new content."

for name, model_id in models.items():
    print(f"\n=== {name.upper()} ===")
    try:
        fill_mask = pipeline("fill-mask", model=model_id)
        output = fill_mask(text)
        for o in output[:3]:
            print(o["sequence"], " | score:", round(o["score"], 4))
    except Exception as e:
        print("ERROR:", e)
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for name, model_id in models.items():
    print(f"\n=== {name.upper()} ===")
    try:
        qa = pipeline("question-answering", model=model_id)
        result = qa(question=question, context=context)
        print("Answer:", result["answer"])
        print("Score:", round(result["score"], 4))
    except Exception as e:
        print("ERROR:", e)
```
